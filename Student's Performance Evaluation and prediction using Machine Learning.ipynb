```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import seaborn as sn
import matplotlib.pyplot as plt

loan_data = pd.read_excel('/content/sample_data/newdataset.xlsx')

# Convert to categorical
loan_data['outcome'] = loan_data['outcome'].astype('category')
loan_data['P'] = loan_data['P'].astype('category')
loan_data['C'] = loan_data['C'].astype('category')
loan_data['TotalBacklogs'] = loan_data['TotalBacklogs'].astype('category')

predictors1 = ['P', 'TotalBacklogs', 'C']
outcome1 = 'outcome'
X1 = pd.get_dummies(loan_data[predictors1], prefix='', prefix_sep='')
y1 = loan_data[outcome1]

# Splitting the dataset
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.25)

naive_model = MultinomialNB(alpha=0.01, fit_prior=True)
naive_model.fit(X_train1, y_train1)

new_loan = X1.loc[146:146, :]
print('Predicted class:', naive_model.predict(new_loan)[0])

probabilities = pd.DataFrame(naive_model.predict_proba(new_loan),
                             columns=naive_model.classes_)
print('Predicted probabilities:')
print(probabilities)

# Assessing the Naive Bayesian model
y_pred1 = naive_model.predict(X_test1)
# Printing the accuracy
print("Accuracy score for Naive Bayesian model:", accuracy_score(y_test1, y_pred1))
# Confusion matrix
cm1 = metrics.confusion_matrix(y_test1, y_pred1)
print("Confusion matrix for Naive Bayesian model:")
print(cm1)

import seaborn as sn
sn.heatmap(cm1, annot=True,  fmt='.2f', xticklabels = ["GATE","HIGHERSTUDIES","NOT ELIGIBLE","PROGRAMMER"] , yticklabels = ["GATE","HIGHERSTUDIES","NOT ELIGIBLE","PROGRAMMER"])
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)

print("METRICS  FOR NAIVE BAYESIAN MODEL ARE:")
print(metrics.classification_report(y_test1,y_pred1,digits=2))

```
